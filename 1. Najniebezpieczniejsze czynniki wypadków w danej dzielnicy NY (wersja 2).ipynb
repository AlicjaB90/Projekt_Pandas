Uzyskanie tabeli z potrzebnymi danymi do przeprowadzenia analizy

```python
nypd_factors= nypd_df[['BOROUGH','CONTRIBUTING FACTOR VEHICLE 1', 'CONTRIBUTING FACTOR VEHICLE 2', 'CONTRIBUTING FACTOR VEHICLE 3', 'CONTRIBUTING FACTOR VEHICLE 4', 'CONTRIBUTING FACTOR VEHICLE 5', 'COLLISION_ID']]
nypd_factors
```

Aby mieć pewność, że dane nie będą się powielać usuniemy duplikaty w kolumnie COLLISION_ID. Kolejnym krokiem jest usunięcie wierszy, które w kolumnie COLLISION_ID nie posiadają wartości. 

```python
nypd_factors_clean = nypd_factors.drop_duplicates(subset=['COLLISION_ID'])
nypd_factors_clean = nypd_factors.dropna(subset=['COLLISION_ID'])
nypd_factors_clean
```

Kolejnym krokiem jest stworzenie nowej kolumny o nazwie FACTORS z połączonymi czynnikami dla danego COLLISION_ID

```python
nypd_factors_clean['FACTORS'] = nypd_factors_clean[['CONTRIBUTING FACTOR VEHICLE 1', 'CONTRIBUTING FACTOR VEHICLE 2', 'CONTRIBUTING FACTOR VEHICLE 3', 'CONTRIBUTING FACTOR VEHICLE 4', 'CONTRIBUTING FACTOR VEHICLE 5']].values.tolist()
nypd_factors_clean
```

Uzyskanie tabeli potrzebnej do dalszej analizy, czyli z nazwami dzielnic i czynnikami w formie listy

```python
nypd_factors_clean=nypd_factors_clean[['BOROUGH', 'FACTORS']]
nypd_factors_clean
```

Za pomocą explode każdy czynnik został przypisany do danej dzielnicy

```python
nypd_factors_clean_exploded = nypd_factors_clean.explode('FACTORS')
nypd_factors_clean_exploded
```

Grupujemy ile razy określony czynnik występował w danej dzielnicy

```python
frequency_counts = nypd_factors_clean_exploded.groupby(['BOROUGH', 'FACTORS']).size().reset_index(name='COUNT')
frequency_counts
```

Sortowanie od czynników, które najczęściej występują w danej dzielnicy

```python
frequency_counts_sort=frequency_counts.sort_values(by='COUNT', ascending=False)
frequency_counts_sort
```

Usuwamy czynniki niezdefiniowane (Unspecified), jeśli celem analizy jest ocena czynników zdefiniowanych

```python
top_factors_specified = nypd_factors_clean_exploded[nypd_factors_clean_exploded['FACTORS'] != 'Unspecified']
top_factors_specified
```

Uzyskanie 3 najniebezpieczniejszych czynników dla danej dzielnicy w NY. 

```python
top_factors_specified = top_factors_specified.groupby('BOROUGH')['FACTORS'].value_counts().groupby(level=0).head(3)
top_factors_specified
```
