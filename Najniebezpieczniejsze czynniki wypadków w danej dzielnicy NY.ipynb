```python
#wczytanie danych

nypd_df = pd.read_csv("/home/student/PythonDeveloper/pythonProject/Projekt_Pandas/data/nypd-motor-vehicle-collisions.csv")#, usecols=["name", "ingredients"]
nypd_df
```

```python
#wydzielenie kolumny z dzielnicami NY
borough_nypd_df= nypd_df[['BOROUGH']]
borough_nypd_df
```

```python
#określenie ile nazw dzielnic występuje
borough_nypd_df.drop_duplicates()
```

```python
#oczyszczenie nazw dzielnic z Nan
nypd_df_clean = nypd_df.dropna(subset=['BOROUGH']).copy()
nypd_df_clean
```

```python
#wydzielenie z df ramki z danymi dotyczącymi dzielnicy Queens
nypd_df_clean_queens = nypd_df_clean[
    nypd_df_clean['BOROUGH'].str.contains('queens', case=False)
]
nypd_df_clean_queens
```

```python
queens_factors=nypd_df_clean_queens.iloc[:,18:23]
queens_factors
```

```python
#określenie jak często dany str występuje określonej kolumnie
queens_factors_numbers = queens_factors[['CONTRIBUTING FACTOR VEHICLE 1','CONTRIBUTING FACTOR VEHICLE 2','CONTRIBUTING FACTOR VEHICLE 3','CONTRIBUTING FACTOR VEHICLE 4','CONTRIBUTING FACTOR VEHICLE 5']].apply(pd.Series.value_counts) 
queens_factors_numbers
```

```python
#określenie ile razy dany czynniki występuje w wszystkich sześciu kolumnach  (zsumowanie) oraz posortowanie od wartości najwyższej i uzyskanie odpowiedzi na zadane pytania
queens_factors_numbers['Sum'] = queens_factors_numbers[['CONTRIBUTING FACTOR VEHICLE 1','CONTRIBUTING FACTOR VEHICLE 2','CONTRIBUTING FACTOR VEHICLE 3','CONTRIBUTING FACTOR VEHICLE 4','CONTRIBUTING FACTOR VEHICLE 5']].sum(axis=1)
queens_factors_numbers.sort_values(by='Sum', ascending=False)
```
