## Przedstawienie 5 najniebezpieczniejszych czynników wypadków w NY

### Wczytywanie danych z pliku CSV

```python
import pandas as pd
import requests

nypd_df = pd.read_csv("/home/student/nypd-motor-vehicle-collisions.csv")
nypd_df
```

##### Wczytanie tabeli ze wszytskimi danymi do analizy przyczyn oraz lokalizacji wypadków w NY

### Wyznaczenie 5 kolumn z ilością niebezpiecznych czynników wypadków

```python
nypd_df_factors_data = nypd_df[['CONTRIBUTING FACTOR VEHICLE 1','CONTRIBUTING FACTOR VEHICLE 2','CONTRIBUTING FACTOR VEHICLE 3','CONTRIBUTING FACTOR VEHICLE 4','CONTRIBUTING FACTOR VEHICLE 5']].apply(pd.Series.value_counts) 
nypd_df_factors_data
```

##### Przedstawienie zsumowanej ilości przyczyn wypadków dla poszczególnego pojazdu ("CONTRIBUTING FACTOR VEHICLE 1,2,3,4,5")

### Resetowanie indeksu i zmiana nazwy kolmny z czynnikami wypadków. Dodanie sumy wypadkow z 5 kolumn z iloscia czynnikow.

```python
df = nypd_df_factors_data.reset_index()
df = df.rename(columns={df.columns[0]: 'CONTRIBUTING FACTORS'})
df['Sum'] = df[['CONTRIBUTING FACTOR VEHICLE 1','CONTRIBUTING FACTOR VEHICLE 2','CONTRIBUTING FACTOR VEHICLE 3','CONTRIBUTING FACTOR VEHICLE 4','CONTRIBUTING FACTOR VEHICLE 5']].sum(axis=1)
df
```

##### Zresetowanie pierwszego indeksu w tabeli, aby był przedstawiony jako kolumna z nazwą przyczyn wypadków. Zsumowanie przyczyn wypadków dla poszczególnych pojazdów 1,2,3,4,5 i dodanie kolumny z ich sumą.

### Usunięcie niepotrzebnych kolumn

```python
df.drop(['CONTRIBUTING FACTOR VEHICLE 1','CONTRIBUTING FACTOR VEHICLE 2','CONTRIBUTING FACTOR VEHICLE 3','CONTRIBUTING FACTOR VEHICLE 4','CONTRIBUTING FACTOR VEHICLE 5'], axis=1, inplace=True)
df
```

##### Przedstawienie tabeli z przyczyną wypadków oraz ich sumą dla poszczególnych czynników

### Posortowanie i przestawienie 5 niebezpieczennych czynników wg ich najczęstszego występowania

```python
df = df.sort_values(by='Sum', ascending=False)
df.head(5)
```
