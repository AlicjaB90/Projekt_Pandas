# Przedstawienie 5 najniebezpieczniejszych czynników wypadków pod względem śmiertelności w NY

## Wczytywanie danych z pliku CSV

```python
import pandas as pd
import numpy as np

import requests

nypd_df = pd.read_csv("/home/student/nypd-motor-vehicle-collisions.csv")
nypd_df
```

##### Wczytanie tabeli ze wszytskimi danymi do analizy czynników wypadków w NY

## Wyznaczenie kolumn z niebezpiecznymi czynnikami wypadków, osób poszkodowanych oraz ich śmiertelnością

```python
Selected_data_columns = nypd_df[['NUMBER OF PERSONS KILLED','CONTRIBUTING FACTOR VEHICLE 1','CONTRIBUTING FACTOR VEHICLE 2','CONTRIBUTING FACTOR VEHICLE 3','CONTRIBUTING FACTOR VEHICLE 4','CONTRIBUTING FACTOR VEHICLE 5']]
Selected_data_columns
```

##### Kolumna z danymi do określenia 5 czynników wypadków oraz ich śmiertelność w NY

## Usunięcie wierszy, które w kolumnach osób zabitych mają nieokreśloną wartość (NaN)

```python
nypd_factors = Selected_data_columns.dropna(subset=['NUMBER OF PERSONS KILLED'])
nypd_factors
```

##### Oczyszczona tabela z niepotrzebnych(nieokreślonych) wartości

## Dodanie kolumny z czynnikiem wypadków

```python
nypd_factors['CONTRIBUTING FACTORS'] = nypd_factors[['CONTRIBUTING FACTOR VEHICLE 1', 'CONTRIBUTING FACTOR VEHICLE 2', 'CONTRIBUTING FACTOR VEHICLE 3', 'CONTRIBUTING FACTOR VEHICLE 4', 'CONTRIBUTING FACTOR VEHICLE 5']].values.tolist()
nypd_factors
```

##### Dodanie kolumny z czynnikiem wypadków, wynikająca z połączenia wartości z 5 kolumn z czynnikami wypadków dla pojazdów 1,2,3,4,5

## Wyznaczenie kolumn potrzebnych do dalszej analizy

```python
df = nypd_factors[['NUMBER OF PERSONS KILLED', "CONTRIBUTING FACTORS"]]
df
```

##### Została utworzona tabela z kolumną osób zabitych oraz czynnikami wypadków 

## Zamiana wartości 0 w kolumnie z osobami zabitymi na wartość NaN, a następnie usunięcie tych wartości, gdyż nie będą one potrzebne do dalszej analizy

```python
df['NUMBER OF PERSONS KILLED'] = df['NUMBER OF PERSONS KILLED'].replace(0, np.nan)
df
```

```python
df = df.dropna(subset=['NUMBER OF PERSONS KILLED'])
df
```

##### Oczyszczona tabela danych z conajmniej jedną ofiarą wypadków w NY

## Przypisanie ofiary do każdego z czynników wypadków

```python
df_exploded = df.explode('CONTRIBUTING FACTORS')
df_exploded
```

##### Za pomocą metody explode, każdy czynnik wypadku został przypisany do ofiary

## Usunięcie wartości NaN z czynników wypadków

```python
df_exploded_drop = df_exploded.dropna()
df_exploded_drop
```

##### Oczyszczona tabela danych z pustych wartości dla czynników wypadków

## Zsumowanie liczby osób zabitych do poszczególnego czynnika wypadku

```python
person_of_killed = df_exploded_drop.groupby(['CONTRIBUTING FACTORS'])['NUMBER OF PERSONS KILLED'].sum()
person_of_killed
```

##### Tabela zawierającą zsumowaną liczbę ofiar dla każdego z czynnika wypadku

## Wyznaczenie 5 czynników, w których najcześciej występują ofiary śmiertelne wypadków w NY

```python
top_data_of_killed = person_of_killed.groupby('CONTRIBUTING FACTORS').nlargest(5).reset_index(level=0, drop=True)
top_data_of_killed = top_data_of_killed.reset_index()

top_data_of_killed.sort_values('NUMBER OF PERSONS KILLED',ascending=False).head(5)
```

##### Tabela przedstawia 5 najczęstsze przyczyny wypadków, w których jest najwięcej ofiar śmiertelnych
